```python
from collections import deque
```

```python
class CQueue:
    def __init__(self):
        self.queue=deque()
    def appendTail(self, value: int) -> None:
        self.queue.append(value)
    def deleteHead(self) -> int:
        if len(self.queue)==0:
            return -1
        else:
            return self.queue.popleft()
```

```python
cq=CQueue()
```

```python
cq.appendTail(3)
```

```python
cq.deleteHead()
```
