```python
from collections import deque
```

```python
class RecentCounter:

    def __init__(self):
        self.deque=deque()

    def ping(self, t: int) -> int:
        while self.deque and self.deque[0]<t-3000:
            self.deque.popleft()
        self.deque.append(t)
        return len(self.deque)
```

```python
obj = RecentCounter()
```

```python
param_1 = obj.ping(9100)
print(param_1)
```
