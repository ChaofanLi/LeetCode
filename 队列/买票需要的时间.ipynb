```python
from collections import deque
```

```python
class Solution:
    def timeRequiredToBuy(self, tickets, k):
        tickets=deque(tickets)
        result=0
        while tickets:
            if k == 0 and tickets[0]==1:
                return result+1
            elif k==0 and tickets[0]>1:
                result+=1
                x=tickets.popleft()
                tickets.append(x-1)
                k=len(tickets)-1
            elif k!=0 and tickets[0]==1:
                result += 1
                tickets.popleft()
                k-=1
            else:
                result+=1
                x=tickets.popleft()
                tickets.append(x-1)
                k-=1 
        return result
```

```python
class Solution:
    def timeRequiredToBuy(self, tickets, k):
        result=0
        for i in range(len(tickets)):
            if i<=k:
                result += min(tickets[i],tickets[k])
            else:
                result += min(tickets[i],tickets[k]-1)
        return result
```
