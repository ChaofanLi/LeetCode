```python
from collections import deque
```

```python
class MovingAverage:

    def __init__(self, size: int):
        """
        Initialize your data structure here.
        """
        self.size=size
        self.count=0
        self.sum=0
        self.window=deque()


    def next(self, val: int) -> float:
        self.count+=1
        if self.count<=self.size:
            self.window.append(val)
            self.sum+=val
            return self.sum/self.count
        else:
            self.count+=1
            self.sum=self.sum-self.window.popleft()+val
            self.window.append(val)
            return self.sum/self.size
```

```python
class MovingAverage:

    def __init__(self, size: int):
        self.size = size
        self.cnt = 0
        self.sum = 0
        self.ls = []
        

    def next(self, val: int) -> float:
        self.ls.append(val)
        self.cnt += 1
        if self.cnt <= self.size:
            self.sum += self.ls[-1]
            return self.sum / self.cnt
        else:
            self.sum += self.ls[-1] - self.ls[-self.size - 1]
            return self.sum / self.size
```
