```python
from collections import deque
```

```python
class Solution:
    def maxTaskAssign(self, tasks, workers, pills, strength):
        tasks.sort()
        workers.sort()
        start=0
        end=min(len(tasks),len(workers))+1
        while start+1<end: # 循环结束标志1
            m=(start+end)//2  # 二分查找
            # print("try",m) # 匹配tasks[:m] 和 works[-m,:]
            i=0
            p=pills
            fail=False
            valid_tasks=deque()
            for j in range(len(workers)-m,len(workers)):
                w=workers[j]
                while i<m and tasks[i]<=w+strength: # 如果tasks[i]可以被工人j吃药完成
                    valid_tasks.append(tasks[i])
                    i+=1
                if not valid_tasks: #tasks中最小的值都比workers[j]+strength大
                    fail=True
                    break
                if valid_tasks[0]<=w: # 不需要药丸即可完成任务,从左边删除
                    valid_tasks.popleft()
                else:
                    if not p:# 没有药丸
                        fail=True
                        break
                    p-=1 # 药丸减一
                    valid_tasks.pop() # 删除可执行的最大任务
            if fail:
                end=m
            else:
                start=m
        return start
```
