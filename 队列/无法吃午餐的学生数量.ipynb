```python
from collections import deque
```

```python
class Solution:
    def countStudents(self, students,sandwiches):
        students=deque(students)
        sandwiches=sandwiches[::-1]
        while len(sandwiches) != 0 and sandwiches[-1] in students:
            if students[0] == sandwiches[-1]:
                sandwiches.pop()
                students.popleft()
            else:
                flag=students.popleft()
                students.append(flag)
        return len(students)
```

```python
# students = [1,1,0,0]
# sandwiches = [0,1,0,1]
students = [1,1,1,0,0,1]
sandwiches = [1,0,0,0,1,1]
```

```python
so=Solution()
so.countStudents(students,sandwiches)
```
