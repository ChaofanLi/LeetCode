## 双指针

```python
class Solution:
    def isPalindrome(self, head):
        ls = []
        while head:
            ls.append(head.val)
            head = head.next
        return ls[::-1] == ls
```

## 递归

```python
class Solution:
    def isPalindrome(self, head):
        self.head_org = head
        def func(lsnode=head):
            if lsnode:
                if not func(lsnode.next):
                    return False
                if self.head_org.val != lsnode.val:
                    return False
                self.head_org = self.head_org.next
            return True
        return func()
```
