```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def getDecimalValue(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: int
        """
        ls=[]
        while head:
            ls.append(head.val)
            head=head.next
        res=0
        for i in range(len(ls)):
            res+=ls[i]*2**(len(ls)-i-1)
        return res
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def getDecimalValue(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: int
        """
        res=0
        while head:
            res=res*2+head.val
            head=head.next
        return res
        
```
