```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeElements(self, head, val):
        vir_lsnode=ListNode(next=head)
        cur=vir_lsnode
        while cur.next:
            if cur.next.val == val:
                cur=cur.next.next
            else:
                cur=cur.next
        return vir_lsnode.next
```
