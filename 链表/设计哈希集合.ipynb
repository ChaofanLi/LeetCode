```python
class MyHashSet:

    def __init__(self):
        self.ls=[]

    def add(self, key: int) -> None:
        if key not in self.ls:
            self.ls.append(key)

    def remove(self, key: int) -> None:
        if key in self.ls:
            self.ls.remove(key)
        
    def contains(self, key: int) -> bool:
        if key in self.ls:
            return True
        else: 
            return False
```
