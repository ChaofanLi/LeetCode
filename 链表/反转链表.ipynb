## 双指针

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head):
        cur, pre = head,None
        while cur:
            temp = cur.next
            cur.next = pre
            pre = cur
            cur = temp
        return pre
```

## 递归

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head):
        def recur(cur, pre):
            if not cur: return pre    
            res = recur(cur.next, cur) 
            cur.next = pre       
            return res   
        return recur(head, None) 
```
