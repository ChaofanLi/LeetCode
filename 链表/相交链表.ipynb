```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA,headB):
        A=headA
        B=headB
        while A != B:
            A = A.next if A.next else B
            B = B.next if B.next else A
        return A
        
```
