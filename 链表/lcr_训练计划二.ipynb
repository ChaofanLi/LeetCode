```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def trainingPlan(self, head: Optional[ListNode], cnt: int) -> Optional[ListNode]:
        fast,slow=head,head
        d=0
        while fast:
            if d <cnt:
                fast=fast.next
                d+=1
            else:
                fast=fast.next
                slow=slow.next
        return slow
        
```
