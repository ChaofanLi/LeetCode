```python
class Solution:
    def rotate(self, matrix) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        flag=[[x for x in ls] for ls in matrix]
        n=len(matrix)
        for i in range(n):
            for j in range(n):
                matrix[j][n-i-1]=flag[i][j]
```
