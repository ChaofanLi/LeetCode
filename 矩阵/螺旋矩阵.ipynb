```python
class Solution:
    def spiralOrder(self, matrix):
        if matrix == []:
            return matrix
        s, x, z, y = 0, len(matrix)-1, 0, len(matrix[0])-1
        res = []
        while True:
            for i in range(z, y+1):
                res.append(matrix[s][i])
            s += 1
            if s > x:
                break
            for i in range(s, x+1):
                res.append(matrix[i][y])
            y -= 1
            if y < z:
                break
            for i in range(y, z-1, -1):
                res.append(matrix[x][i])
            x -= 1
            if x < s:
                break
            for i in range(x, s-1, -1):
                res.append(matrix[i][z])
            z += 1
            if z > y:
                break
        return res
```

```python
matrix=[[1,2,3],[4,5,6],[7,8,9]]
```

```python
solution=Solution()
solution.spiralOrder(matrix)
```
